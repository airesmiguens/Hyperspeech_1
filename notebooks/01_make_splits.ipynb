# 01_make_splits
Creates leakage-safe StratifiedGroupKFold splits and saves them once.

```python
from pathlib import Path
import sys
import numpy as np

ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.data import load_dataset
from src.splits import make_outer_splits, save_splits

bundle = load_dataset(
    csv_path=ROOT / "data" / "all_data_clean.csv",
    target_col="sbp_binary",
    subject_col="PAT_ID",
)

y = bundle.df[bundle.target_col].to_numpy(dtype=int)
groups = bundle.df[bundle.subject_col].to_numpy()
splits = make_outer_splits(y=y, groups=groups, n_splits=5, seed=42)

out_path = ROOT / "artifacts" / "splits" / "outer_splits.json"
save_splits(splits, out_path)
print(f"Saved {len(splits)} outer folds to {out_path}")
```
